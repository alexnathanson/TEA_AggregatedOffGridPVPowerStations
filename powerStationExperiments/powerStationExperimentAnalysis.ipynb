```python
import pandas as pd
import matplotlib.pyplot as plt  # for visualization
from datetime import datetime, timedelta
```

```python
#convert time strings to datetime objects
exp1_df['time'] = exp1_df['time'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S.%f'))
```

```python
plt.plot(exp1_df['time'],exp1_df['battery'], color='r')
plt.title('Battery Percentage')
plt.xlabel('Time')
plt.ylabel('Percentage')
plt.show() 
```

```python
# get self consumption percentage rate per hour
tDif = exp1_df['time'].iloc[-1]-exp1_df['time'].iloc[0]
pDif = exp1_df['battery'].iloc[0]-exp1_df['battery'].iloc[-1]

avgHourlyPercentage = pDif/(tDif / timedelta(hours=1))
avgHourlyW = 1152 * avgHourlyPercentage * .01
avgHourlyW
```
